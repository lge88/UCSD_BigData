##assigning id using self-defined class##

```python
# define the namespace of ids
import random
from sets import Set

class NameSpace:
    def __init__(self):
        self.generator = lambda : random.randint(1,10000)
        self.used_names = Set()
    
    def getNewId(self):
        tmp_id = str(self.generator())
        while tmp_id in self.used_names:
            tmp_id = str(self.generator())
        
        self.used_names.add(tmp_id)
        return tmp_id
    
    def removeId(self, existing_id):
        if existing_id in self.used_names:
            self.used_names.remove(existing_id)

id_generator = NameSpace()
for single_leaf in leaves:
    single_leaf.id = id_generator.getNewId()
    
# output "old_group.txt" for comparison
with open("old_group2.txt", "wb") as f:
    for node in leaves:
        station = node.stationList
        for single_station in station:
            f.write(single_station + " " + node.id + "\n")
```

## mesh creation ##

```python
from math import *

# step 0: define a class that can represent each cell
class Cell:
    def __init__(self):
        self.cid = None   # cell id would the same as the leaf_node id that contains it

# step 1: generate a mesh
#   input: the world coordinates
#   output a m * n matrix, containing objects called cell

df_index = station_dict.index

def find_min_max_lat_log(leaf):
    up = 0
    down = 0
    left = 0
    right = 0
    
    useless = True
    for single_station in leaf.stationList:
        
        if single_station not in df_index: 
            #print "skipping:" + single_station
            continue  
            
        useless = False
        lat = station_dict.loc[single_station].latitude
        log = station_dict.loc[single_station].longitude
        up = lat if lat > up else up
        down = lat if lat < down else down
        left = log if log < left else left
        right= log if log > right else right
    
    if useless == True:
        return None, None, None, None
    
    # change the cordinates
    up = -up + 90
    down = -down + 90
    left = left + 180
    right = right + 180
    
    return up, left, down, right
    
def create_world_mesh(leaves):
    min_length_lat = 180
    min_length_log = 360

    for leaf in leaves:
        min_lat, min_log, max_lat, max_log = find_min_max_lat_log(leaf)
        if min_lat != None:
            if max_lat - min_lat < min_length_lat: min_length_lat = max_lat - min_lat
            if max_log - min_log < min_length_log: min_length_log = max_log - min_log


    SCALING = 0.5
    incre_lat = min_length_lat * SCALING
    incre_log = min_length_log * SCALING
    incre = incre_lat if incre_lat < incre_log else incre_log
    
    row = int(floor(180 / incre))
    col = int(floor(360 / incre))

    mesh = [[Cell() for x in xrange(col)] for x in xrange(row)]
    return mesh, incre

# call the function
mesh, incre = create_world_mesh(leaves)
```

```python
# step 2: traverse all the leaves of the partition tree, extract the log/lat for corners
# convert to coordinates, and mark the m * n matrix with the group name
for i in range(0, len(leaves)):
    min_lat, min_log, max_lat, max_log = find_min_max_lat_log(leaves[i])
    
    if min_lat == None: continue
    
    startX = int(floor(min_lat / incre))
    endX   = int(floor(max_lat / incre))
    startY = int(floor(min_log / incre))
    endY   = int(floor(max_log / incre))
    
    for j in range(startX, endX + 1):
        for k in range(startY, endY + 1):
            x = j if j < len(mesh) else len(mesh) - 1
            y = k if k < len(mesh[0]) else len(mesh[0]) - 1
            
            mesh[x][y].cid = leaves[i].id # assign the tree node id to matrix cell
```

```python
# in the case of a real map, coordinates might not be such ideal, and there 
# might be cells in the matrix with no id yet. so we do some processing before continue
for i in range(0, len(mesh)):
    last_id = None
    for j in range(0, len(mesh[0])):
        if mesh[i][j].cid != None: last_id = mesh[i][j].cid
        else:
            if last_id == None:
                all_none = True
                for k in range(j, len(mesh[0])):
                    if mesh[i][k].cid !=None: 
                        last_id = mesh[i][k].cid
                        all_none = False
                        break
                
                # the entire line does not have an id
                if all_none == True: continue # we go to the next line
                
            # assign last_id to current
            mesh[i][j].cid = last_id
```

## create the graph structure ##

```python
class GraphNode:
    def __init__(self):
        self.neighbors  = Set()
        self.id = None
        self.length = None

# generate nodes representing each partition
id_to_node = {}
for leaf in leaves:
    gNode = GraphNode()
    gNode.gid = str(leaf.id)
    gNode.station = leaf.stationList

    # add to the dict
    id_to_node[gNode.gid] = gNode


# connecting the graph nodes, with the help of mesh matrix
x_offset = [0, 1, 0, -1]
y_offset = [-1,0, 1, 0 ]

for i in range(0, len(mesh)):
    for j in range(0, len(mesh[0])):
        curNode_id = mesh[i][j].cid
        if curNode_id == None: continue
        
        curNode = id_to_node[curNode_id]
        
        for k in range(0, 4):
            new_x = i + x_offset[k]
            new_y = j + y_offset[k]
            
            # out of bound check
            if new_x < 0 or new_x > len(mesh)-1 or new_y < 0 or new_y >len(mesh[0])-1: continue
                
            # id == None check
            newNode_id = mesh[new_x][new_y].cid
            if newNode_id == None: continue

            # 
            
            newNode = id_to_node[newNode_id]
            if newNode_id != curNode_id and newNode not in curNode.neighbors:
                curNode.neighbors.add(newNode)
```

##recursively merge node ##


repeat this step until at least one of the following condition holds
   1) nothing to merge
   2) the max_iteration has been reached.


```python
# would need a station_to_mm structure
station_to_mm = {}

def get_data_from_string(line):
    tmp_arr = line.split(',')
    ret_arr = []
    for item in tmp_arr:
        ret_arr.append(int(item))
    return ret_arr
    
with open("mapping.txt", "rb") as f:
    for line in f:
        line_sp = line.split()
        station_name = line_sp[0][1:-1]
        data = get_data_from_string(line_sp[1][1:-1]) 
        station_to_mm[station_name] = data
```

```python
%time
import numpy as np

def find_length(curNode):
    # use curnode.station and the lookup table for: station_name -> max_min data
    
    station = curNode.station
    #working_list = station[0:100] if len(station) >= 100 else station
    working_list = station
    
    # find the mean
    mean = [0] * 365
    count = 0
    for itr in working_list:
        if itr in station_to_mm:
            tmp1 = station_to_mm[itr][:365]
            tmp2 = station_to_mm[itr][365:]
            mean_tmp = np.divide(np.add(tmp1, tmp2), 2)
            count += 1
            mean = np.add(mean, mean_tmp)
    count = 1 if count == 0 else count
    mean = np.divide(mean, count)

    # find the sum
    mat_sum = np.zeros([365, 365])
    for itr in working_list:
        if itr in station_to_mm:
            tmp1 = station_to_mm[itr][:365]
            tmp2 = station_to_mm[itr][365:]
            mean_tmp = np.divide(np.add(tmp1, tmp2), 2)
            
            data = np.subtract(mean_tmp, mean)
            mat_tmp = np.outer(data, data)  #should use outer, instead of multiply!
            mat_sum = np.add(mat_sum, mat_tmp)

    mat_sum = np.divide(mat_sum, count)

    # do the svd
    U, D, V = np.linalg.svd(mat_sum)
    D_scale = D[:]
    try:
        acc_sum = np.cumsum(D_scale[:])
        acc = np.divide(acc_sum, sum(D_scale))

        REQUIREMENT = 0.95
        i = 0
        for i in range(0, len(acc)):
            if acc[i] >= REQUIREMENT: break 
        
        k = i + 1
        n = len(working_list)
        
        #return n*k + (k+1)*730
        return n*k + (k+1)*365
        
    except Exception:
        print "error extracting k"
        return -1

def merge_two_nodes(node1, node2):
    tmp_node = GraphNode()
    tmp_node.station = node1.station + node2.station # appending two list together
    l = find_length(tmp_node)
    if l == -1: return None
    
    l1 = node1.length if node1.length != None else find_length(node1)
    l2 = node2.length if node2.length != None else find_length(node2)
    if l1 == -1 or l2 == -1: return None
    
    tmp_node.length = l
    node1.length = l1
    node2.length = l2
    
    if l <= l1 + l2 : # should merge
        tmp_node.neighbors = node1.neighbors.union(node2.neighbors)
        tmp_node.gid = node1.gid
        id_generator.removeId(node2.gid)
        
        print tmp_node.gid, node1.gid, node2.gid
        print l, l1, l2
        print len(tmp_node.station), len(node1.station), len(node2.station)
        print ""
        
        for each_node in tmp_node.neighbors:
            if node1 in each_node.neighbors: each_node.neighbors.remove(node1)
            if node2 in each_node.neighbors: each_node.neighbors.remove(node2)
            each_node.neighbors.add(tmp_node)
        
        return tmp_node
    else:
        return None
  

# the routine for doing partition
active_nodes = Set(id_to_node.values()) # the return type is a list, and we can use in/remove etc
fixed_nodes  = Set()
MAX_STEPS    = 3000
i = 0

while len(active_nodes) != 0 and i < MAX_STEPS:
    curNode = active_nodes.pop()
    neib = curNode.neighbors

    merged = False
    for node in neib:
        if node not in active_nodes: continue
        new_node = merge_two_nodes(curNode, node)
        
        if new_node != None:
            active_nodes.remove(node)
            active_nodes.add(new_node)
            merged = True
            break

    if merged == False:
        fixed_nodes.add(curNode)
        
    # increment the step count
    i += 1
```

```python
# create the old leaves
def getList(file_name):
    retList = []
    id_to_node = {}
    
    with open(file_name,"rb") as f:
        for line in f:
            arr = line.split()
            station_name = arr[0]
            node_id =  arr[1]
            
            if node_id in id_to_node:
                node = id_to_node[node_id]
                node.stationList.append(station_name)
            else:
                new_node = TreeNode()
                new_node.id = node_id
                new_node.stationList = [station_name]
                id_to_node[node_id] = new_node
                retList.append(new_node)
    return retList

old_group = getList("old_group2.txt")
new_group = getList("new_group2.txt")
```
